```python
from tensorflow.keras.preprocessing.text import one_hot
import numpy as np
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### One Hot Representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]  ##converted to vectors
one_hot_repr
```

```python
## word Embedding Representation

from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation
dim=10   ##10 features
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length)) ##voc_size,dim,input_length are the parameters for embedding layer(word2vec)
model.compile('adam','mse') ##compiling the model
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
